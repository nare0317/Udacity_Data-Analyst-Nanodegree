#### Full state names sometimes, abbreviations other times

##### Define


##### Code

```python
# Check value counts for "state" column which have full name
patients_clean[patients_clean.state.str.len() != 2].state.value_counts()
```

```python
# Mappiing from full state name to abbreviations in those 5 values
state_abb = {'California' : 'CA', 
             'New York' : 'NY', 
             'Illinois': 'IL',
             'Florida': 'FL',
             'Nebraska' : 'NE'
            }

# Make a function to apply 
def full_to_abb(patient):
    if patient['state'] in state_abb.keys(): 
        abb = state_abb[patient['state']]
        return abb
    else:
        return patient['state']

# Apply the function 
patients_clean['state'] = patients_clean.apply(full_to_abb, axis=1)
```
